## Feature Understanding

```python
df['Crash_Year'].value_counts()
```

```python
year_order = []

for i in range(2001,2022):
    year_order.append(i)

print(year_order)
```

```python
ax = df['Crash_Year'].value_counts() \
    .loc[year_order] \
    .plot(kind='bar', title="QLD Crash Count by Year")
ax.set_xlabel('Year')
ax.set_ylabel('Count')
```

Upon reviewing the dataset notes it was found that "Property damage only crashes ceased to be reported/recorded by Queensland Police Service after 31 December 2010." Therefore, this has been dropped from the data set.

```python
df.query('Crash_Severity == "Property damage only"')
```

```python
df = df[df.Crash_Severity != 'Property damage only'] \
    .reset_index(drop=True).copy()
```
